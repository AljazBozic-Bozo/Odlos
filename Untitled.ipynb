### Branje ocen

Razred za branje datoteke ocen. <br>
V konstruktor prejme pot do datoteke, od katerega datuma in do katerega datuma upoštevamo ocene ter minimalno število ocen za filme kateri pridejo v poštev.<br>
Da datoteko preberemo moramo uporabiti metodo read().<br>
Za število ocen uporabimo metodo nratings().<br>
Dodal sem tudi metodo calculate katera ustvari dataframe vseh uporabnikov in sicer: njihov ID, koliko filmov so ocenili in povprečno oceno. Df se hrani v razredu kot atribut allUsers.

```python
class UserItemData:
    path = ""
    start_date = ""
    end_date = ""
    min_ratings=0
    
    def __init__(self, path, start_date=None, end_date=None, min_ratings=None):
        self.path = path
        self.start_date = start_date
        self.end_date = end_date
        self.min_ratings=min_ratings
    
    def read(self):
        self.ratings = pd.read_table(self.path, parse_dates={"date":["date_day", "date_month", "date_year"]}, encoding = "ISO-8859-1")
        
        if self.start_date != None:
            self.ratings = self.ratings[self.ratings.date>=self.start_date]
        if self.end_date != None:
            self.ratings = self.ratings[self.ratings.date>=self.end_date]
        if self.min_ratings != None:
            self.ratingsCount = self.ratings["movieID"].value_counts() >= self.min_ratings
            self.ratingsCount = self.ratingsCount[self.ratingsCount==True]
            self.ratings = self.ratings[self.ratings.movieID.isin(self.ratingsCount.keys().tolist())]
    
    def calculate(self):
        ser = self.ratings["userID"].value_counts()

        allUsers = pd.DataFrame({"userID":ser.index, "count":ser.values})
        means = []
        for i in allUsers["userID"]:
            filt = self.ratings["userID"] == i
            temp = self.ratings[filt]
            means.append(temp["rating"].mean())
        allUsers["avg"] = means
        self.allUsers = allUsers
    
    def nratings(self):
         return len(self.ratings)
    
    def ratings(self):
        return self.ratings;
```

### Branje filmov

Razred za branje filmov - MovieData. <br>
Da preberemo datoteko kličemo metodo read().<br>
Za izpis imena filma z ID uporabimo metodo get_title(ID).

```python
class MovieData:
    path = ""
    def __init__(self, path):
        self.path = path
    
    def read(self):
        self.movies = pd.read_table(self.path, encoding = "ISO-8859-1")
        self.len = self.movies.shape[0]
        
    def get_title(self, mid):
        getById = self.movies["id"]==int(mid)
        
        return (self.movies[getById]["title"].values[0])
    
    
```

### Naključni prediktor

Razred za naključno določanje filmov - RandomPredictor <br>
Z metodo predict določimo naključno oceno za vse filme in jih vrnemo v slovarju.

```python
type(md.movies["id"])
```

```python
class RandomPredictor:
    def __init__(self, min_rating, max_rating):
        self.min = min_rating;
        self.max = max_rating;
        
        
    def fit(self, X):
        self.X = X
    
    def predict(self, user_id, md):
        dicto = {}
        movies = md.movies["id"]
        #print((self.X.ratings))
        
        for i in movies:
            dicto[i] = random.randint(self.min, self.max)
        return dicto
    
```

### Priporočanje

Razred za priporočanje - Recommender

```python
class Recommender:
    def __init__(self, prediktor):
        self.pr = prediktor
        
    def fit(self, X):
        self.pr.fit(X)
        self.X = self.pr.X
        self.ratings = self.X.ratings[["userID", "movieID", "rating"]]
    
    def recommend(self, user_id, n=10, rec_seen=True):
        assigned = []
        res = []
        pred = self.pr.predict(user_id, md)
        if not rec_seen:
            filt = (self.ratings["userID"] != user_id)
            self.ratings = self.ratings[filt]
        for index, row in self.ratings.iterrows():
            movID = row["movieID"]
            if movID not in assigned:
                assigned.append(movID)
                #print([movID, pred[movID]])
                res.append([movID, pred[movID]])
            
        res.sort(key=lambda x: x[1])
        return res[-n:]
       
```

Uporaba Recommender

```python
md = MovieData('data/movies.dat')
md.read()
uim = UserItemData('data/user_ratedmovies.dat')
uim.read()
rp = RandomPredictor(1, 5)
rec = Recommender(rp)
rec.fit(uim)
rec_items = rec.recommend(78, n=10, rec_seen=False)

for idmovie, val in rec_items:
    print("Film: {}, ocena: {}".format(md.get_title(idmovie), val))  
```

### Napovedovanje s povprečjem

Razred za napovedovanje s povprečenjem - AveragePredictor <br>
Razred v metodi predict izračuna oceno za vsak film s pomočjo formule avg = (vs + self.b * g_avg) / (n + self.b)
kjer: 
<ul>
    <li>vs je vsota vseh ocen za ta film,</li>
    <li>n je število ocen, ki jih je ta film dobil,</li>
    <li>g_avg je povprečje čez vse filme,</li>
    <li>b je parameter formule za povprečje. Če je b=0, gre za navadno povprečje.</li>
</ul>

```python
class AveragePredictor:
    def __init__(self, b):
        self.b = b
            
    def fit(self, X):
        self.X = X
        self.ratings = X.ratings
    
    def predict(self, user_id, md):
        movies = md.movies
        dicto = {}
        #povprečje čez vse filme
        g_avg = self.ratings["rating"].mean()
        #b zaupanje formuli
        b = self.b
        
        for i in movies["id"]:
            #samo ratingi tega filma
            filt = (self.ratings["movieID"] == i)
            movRatings = self.ratings[filt]
            #vsota ocen za ta film
            vs = movRatings["rating"].sum()
            #število ocen za ta film
            n = movRatings["rating"].shape[0]
            if(not n == 0): dicto[i] = (vs + self.b * g_avg) / (n + self.b)
            else: dicto[i] = 0
            
            
        return dicto
```

Uporaba AveragePredictor z b=100

```python
md = MovieData('data/movies.dat')
md.read()
uim = UserItemData('data/user_ratedmovies.dat')
uim.read()
ap = AveragePredictor(b=100)
rec = Recommender(ap)
rec.fit(uim)
rec_items = rec.recommend(78, n=10, rec_seen=False)
rec_items.reverse()
for idmovie, val in rec_items:
    print("Film: {}, ocena: {}".format(md.get_title(idmovie), val))  
```

### Priporočanje najbolj gledanih filmov

Razred za priporočanje najbolj gledanih filmov - ViewsPredictor <br>
V metodi predict preprosto preštejemo število ogledov(ocen) vsakega filma in to je njegova končna ocena

```python
class ViewsPredictor:
    def __init__(self):
        self.X = 0
        
    def fit(self, X):
        self.X = X
        self.ratings = X.ratings
    
    def predict(self, user_id, md):
        dicto = {}
        movies = md.movies
        
        for i in movies["id"]:
            filt = (self.ratings["movieID"] == i)
            movRatings = self.ratings[filt]
        
            n = movRatings["rating"].shape[0]
            #print([i, n])
            dicto[i] = n
            
        
        return dicto
```
